```python
from sqlalchemy import create_engine
import pandas as pd
import urllib.parse

# Step 1: Encode the password if it contains special characters
password = urllib.parse.quote("deepthi16")  # '@' becomes '%40'

# Step 2: Define connection parameters
username = 'root'
host = 'localhost'
port = 3306
database = 'jobs'

# Step 3: Create the SQLAlchemy connection string
connection_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"

# Step 4: Create the engine
engine = create_engine(connection_string)

# Step 5: Query the table and read into a DataFrame
try:
    query = "SELECT * from job_data;"  # Change to your table
    df = pd.read_sql(query, engine)
    print("✅ Connection successful! Here's your data:")
    display(df)

except Exception as e:
    print(f"❌ Error occurred: {e}")
```

## BI VARIATE

```python
top_locations = df['Location'].value_counts().head(10).index
top_jobs = df['Job'].value_counts().head(10).index

filtered_df = df[df['Location'].isin(top_locations)]
final_df = filtered_df[filtered_df['Job'].isin(top_jobs)]

worktype_by_location = pd.crosstab(final_df['Location'], final_df['Job'])
a = sns.heatmap(worktype_by_location, annot = True)
for bars in a.containers:
    a.bar_label(bars)
plt.title("Top 10 Jobs Distribution in Top 10 Locations")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
