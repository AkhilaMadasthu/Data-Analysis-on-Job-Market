```python
from sqlalchemy import create_engine
import pandas as pd
import urllib.parse

# Step 1: Encode the password if it contains special characters
password = urllib.parse.quote("deepthi16")  # '@' becomes '%40'

# Step 2: Define connection parameters
username = 'root'
host = 'localhost'
port = 3306
database = 'jobs'

# Step 3: Create the SQLAlchemy connection string
connection_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"

# Step 4: Create the engine
engine = create_engine(connection_string)

# Step 5: Query the table and read into a DataFrame
try:
    query = "SELECT * from job_data;"  # Change to your table
    df = pd.read_sql(query, engine)
    print("✅ Connection successful! Here's your data:")
    display(df)

except Exception as e:
    print(f"❌ Error occurred: {e}")
```

```python
df_exploded = df.copy()
df_exploded['Skills'] = df_exploded['Skills'].str.split(',')
df_exploded = df_exploded.explode(['Skills'])
df_exploded['Skills'] = df_exploded['Skills'].str.strip()
```

```python
location_skills = df_exploded.groupby(['Location', 'Skills']).size().unstack(fill_value=0)
top_skills = df_exploded['Skills'].value_counts().head(10).index
location_top_skills = location_skills[top_skills]
location_top_skills.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title("Skills v/s Location")
plt.xlabel("Location")
plt.ylabel("Skill Count")
plt.legend(title="Skills", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
